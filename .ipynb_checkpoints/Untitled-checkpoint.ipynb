## Clusterize numericalized category. 

Apply it for ORGANIZATION_TYPE. 

```python
def category_information(data, col_name):
    name = col_name
    data_train = data.copy()
    collection_data = collections.Counter(data_train[name])
    rate_list = []
    data_list = list(collection_data.keys())
    data_len = [collection_data[k] for k in data_list]
    for element in data_list:
        for_rate = data_train[data_train[name] == element].TARGET
        if len(for_rate) == 0:
            for_rate = data_train[data_train[name].isnull()].TARGET
            rate = for_rate.sum() / len(for_rate)
        else:
            rate = for_rate.sum() / len(for_rate)
        rate_list.append(rate)
    df = pd.DataFrame()
    df['rate'] = rate_list
    df['element'] = [str(key) for key in data_list]
    df['length'] = data_len                                     
    df = df.sort_values('rate', ascending=False).reset_index(drop=True)
    rcParams['figure.figsize'] = 13, 4
    plt.bar(df.element, df.rate)
    plt.ylim(df.rate.min()-df.rate.mean()/10)
    plt.ylabel('rate')
    return df
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
org_df = category_information(data_train, 'ORGANIZATION_TYPE')

linkage_result = linkage(org_df[['rate']], method='ward', metric='euclidean')
#plt.figure(num=None, figsize=(16, 9), dpi=200, facecolor='w', edgecolor='k')
#dendrogram(linkage_result)
fcluster(linkage_result, 0.01, criterion='distance')

org_df['class'] = fcluster(linkage_result, 0.01, criterion='distance')

b_c = np.nan
num = 0
new_class = []

for k, c in enumerate(org_df['class']):
    if c == b_c:
        b_c = c
    else:
        num = num + 1
        b_c = c
    new_class.append(num)

org_df['n_class'] = new_class
```

```python
data_train['ORGANIZATION_TYPE'] = [org_df[org_df['element'] == cam].n_class.values[0]
                                   for cam in data_train['ORGANIZATION_TYPE']]
```

## One hot encoding for category data.

```python
def one_hot_encoding(data, column, drop_first):
    dataset = data.copy()
    add_df = pd.get_dummies(dataset[column], drop_first=drop_first)
    new_col = [column + '_' + str(col) for col in add_df.columns]
    add_df.columns = new_col
    print(add_df.columns)
    del dataset[column]
    return pd.concat([dataset, add_df], axis=1)

def recursive_one_hot(data, column_list, drop_first):
    while len(column_list) > 0 :
        data = one_hot_encoding(data, column_list[0], drop_first=drop_first)
        column_list.pop(0)
        return recursive_one_hot(data, column_list, drop_first=drop_first)
    return data
```

```python
# to bool list.
single_element_list = ['NAME_CONTRACT_TYPE', 'CODE_GENDER' , 'FLAG_OWN_CAR', 'FLAG_OWN_REALTY']

# to one hot list.
multi_element_list = ['NAME_FAMILY_STATUS', 'WEEKDAY_APPR_PROCESS_START']
```

```python
# Create new features.
dataset['NEW_CREDIT_TO_ANNUITY_RATIO'] = dataset['AMT_CREDIT'] / dataset['AMT_ANNUITY']
dataset['NEW_CREDIT_TO_GOODS_RATIO'] = dataset['AMT_CREDIT'] / dataset['AMT_GOODS_PRICE']
dataset['NEW_INC_PER_CHLD'] = dataset['AMT_INCOME_TOTAL'] / (1 + dataset['CNT_CHILDREN'])
dataset['NEW_EMPLOY_TO_BIRTH_RATIO'] = dataset['DAYS_EMPLOYED'] / dataset['DAYS_BIRTH']
dataset['NEW_ANNUITY_TO_INCOME_RATIO'] = dataset['AMT_ANNUITY'] / (1 + dataset['AMT_INCOME_TOTAL'])
dataset['NEW_SOURCES_PROD'] = dataset['EXT_SOURCE_1'] * dataset['EXT_SOURCE_2'] * dataset['EXT_SOURCE_3']
dataset['NEW_EXT_SOURCES_MEAN'] = dataset[['EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3']].mean(axis=1)
dataset['NEW_SCORES_STD'] = dataset[['EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3']].std(axis=1)
dataset['NEW_SCORES_STD'] = dataset['NEW_SCORES_STD'].fillna(dataset['NEW_SCORES_STD'].mean())
dataset['NEW_CAR_TO_BIRTH_RATIO'] = dataset['OWN_CAR_AGE'] / dataset['DAYS_BIRTH']
dataset['NEW_CAR_TO_EMPLOY_RATIO'] = dataset['OWN_CAR_AGE'] / dataset['DAYS_EMPLOYED']
dataset['NEW_PHONE_TO_BIRTH_RATIO'] = dataset['DAYS_LAST_PHONE_CHANGE'] / dataset['DAYS_BIRTH']
dataset['NEW_PHONE_TO_BIRTH_RATIO_EMPLOYER'] = dataset['DAYS_LAST_PHONE_CHANGE'] / dataset['DAYS_EMPLOYED']
dataset['NEW_CREDIT_TO_INCOME_RATIO'] = dataset['AMT_CREDIT'] / dataset['AMT_INCOME_TOTAL']
```
